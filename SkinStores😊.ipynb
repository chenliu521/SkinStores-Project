## Import Package


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import json
```

```python
def single_page_scraper(url):
    Request=requests.get(url=url)
    soup=BeautifulSoup(Request.content)
    return soup

def get_all_hrefs(soup):
    hrefs=[]
    for a in soup.find_all('a',href=True):
        hrefs.append(a['href'])
    return hrefs
```

## Extract all keywords

```python
import re
def extract_keywords(product_details_str):
    splited_text=re.split(",|[|]|/| ",product_details_str)
    one_row_keywords=[]
    for i in splited_text:
        if ":" in i:
            one_row_keywords.append(i)
    return one_row_keywords


all_keywords=[]
for i in range(len(Pro_desc)):
    new_ele=str(Pro_desc[i]).replace("\\n"," ")
    ele_keywords=extract_keywords(new_ele)
    all_keywords.extend(ele_keywords)
    
    
```

```python

def replace_word(array):
    new_array=[]
    for i in array:
        i=i.replace("Description",";Description:")
        i=i.replace("Product Details",";Product Details:")
        i=i.replace("Directions of Use:",";Directions of Use:")
        i=i.replace("Tolerance:",";Tolerance:")
        i=i.replace("Volume:",";Volume:")
        i=i.replace("Directions:",";Directions:")
        i=i.replace("Ingredients:",";Ingredients:")
        i=i.replace("Brand:",";Brand:")
        i=i.replace("Range:",";Range:")
        i=i.replace("Size:",";Size:")
        i=i.replace("Volume:",";Volume:")
        new_array.append(i)
    return new_array
```

```python
new_Pro_desc=[]
for i in Pro_desc:
    new_Pro_desc.append(replace_word(i))
```

```python
new_Pro_desc[0]
```

```python
split_Pro_desc=[]
for i in new_Pro_desc:
    j=str(i).replace("[","").replace("]","").replace("'","").split(";")
    split_Pro_desc.append(j)
    
```

```python
final_keywords=['Brand:',
 'Ingredients:',
 'Directions:',
 'Volume:',
 'Range:',
 'Size:',
 'Benefits:',
 'Description:',
 'Product Details:',"Tolerance:","Directions of Use:"]
```

```python
def data_extraction(keywords,array):
    for i in array:
        if keywords in i:
            return i
```

```python
def get_column(keywords):
    lst=[]
    for i in range(len(split_Pro_desc)):
        desc=data_extraction(keywords,split_Pro_desc[i])
        lst.append(desc)
    return lst
    
```

```python
desc=get_column('Description:')
detials=get_column('Product Details:')

brand=get_column('Brand:')
ingrediants=get_column('Ingredients:')
directions=get_column('Directions:')
volume=get_column('Volume:')
ranges=get_column('Range:')
size=get_column('Size:')
benefits=get_column('Benefits:')
tolerance=get_column("Tolerance:")
directions_use=get_column("Directions of Use:")
```

```python
df=pd.DataFrame()
df["product link"]=all_products_links
df["Description"]=[str(i).replace("\\n","").strip().replace("  ","") for i in desc]
df["detials"]=[str(i).replace("\\n","").strip().replace("  ","")  for i in detials]
df["Brand"]=[str(i).replace("\\n","").strip().replace("  ","")  for i in brand]
df["Ingredients"]=[str(i).replace("\\n","").strip().replace("  ","")  for i in ingrediants]
df["Directions"]=[str(i).replace("\\n","").strip().replace("  ","")  for i in directions]
df["Volume"]=[str(i).replace("\\n","").strip().replace("  ","")  for i in volume]
df["Range"]=[str(i).replace("\\n","").strip().replace("  ","")  for i in ranges]
df["Size"]=[str(i).replace("\\n","").strip().replace("  ","")  for i in size]
df["Benefits"]=[str(i).replace("\\n","").strip().replace("  ","")  for i in benefits]
df["Tolerance"]=[str(i).replace("\\n","").strip().replace("  ","")  for i in tolerance]
df["Directions_Use"]=[str(i).replace("\\n","").strip().replace("  ","")  for i in directions_use]
```
